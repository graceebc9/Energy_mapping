```python
stor = {}
for f in onsud_dir:
    label = f.split('/')[-1].split('.')[0].split('_')[-1]   
    df = pd.read_csv(f)
    pcs = df['PCDS'].unique()    
    stor[label] = pcs 
    
```

```python
ovr_list = [] 
for k, v in stor.items(): 
    for other_k, other_v in stor.items():
        if k != other_k:
            v = set(v)
            other_v = set(other_v)
            overlapping = v.intersection(other_v)
            if overlapping:
                # loop over overalpping and add eahc row seperatly 
                for o in overlapping:
                    ovr_list.append((k, other_k, o, 1))
                


# convert into df÷
df = pd.DataFrame(ovr_list, columns=['file1', 'file2', 'overlapping_postcodes', 'num'])
            
```

```python
print('num of overlapping postcodes ', len(df.overlapping_postcodes.unique()) ) 

```

```python
# convert set into list 
df['overlapping_postcodes'] = df['overlapping_postcodes'].apply(list)

# extract all pcs from table 
all_pcs = []
for k, v in stor.items():
    all_pcs.extend(v)
```
